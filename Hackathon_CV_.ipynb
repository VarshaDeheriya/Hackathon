```python
train_dir = '/content/Intel Image Classification/seg_train/seg_train'
test_dir = '/content/Intel Image Classification/seg_test'

categories = {'buildings': 0, 'forest': 1, 'glacier': 2, 'mountain': 3, 'sea': 4, 'street': 5}
train = []
test = []

# train data
for category, label in categories.items():
    path = os.path.join(train_dir, category)
    for img in os.listdir(path):
        train.append(['train/{}/{}'.format(category, img), label])

# test data
for category, label in categories.items():
    path = os.path.join(test_dir, category)
    for img in os.listdir(path):
        test.append(['test/{}/{}'.format(category, img), label])

df1 = pd.DataFrame(train, columns=['path', 'label'])  # train
df2 = pd.DataFrame(test , columns=['path', 'label'])  # test
```

```python
IMG_SIZE = 100    # image size

categories = ['buildings', 'forest', 'glacier', 'mountain', 'sea', 'street']  # image categories

df1_path = '/content/Intel Image Classification/seg_train/seg_train'   # train path

# pre-process the training data
X_train = []
y_train = []

for category in categories:
    path = os.path.join(df1_path, category)
    class_num = categories.index(category)
    for img in os.listdir(path):
        img_array = np.array(Image.open(os.path.join(path, img)).convert('L').resize((IMG_SIZE, IMG_SIZE)))
        img_array = img_array.flatten()
        img_array = img_array / 255.0
        X_train.append(img_array)
        y_train.append(class_num)

X_train = np.array(X_train)
y_train = np.array(y_train)
```
